### Exercise 2: Check Even or Odd

Write pseudocode that checks if a number is even or odd and prints the result. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
ALGORITHM CheckEvenOrOdd
    // 1. Start the process
    START
    
    // 2. Define the number to check
    READ Number = 7
    
    // 3. Check if the remainder of division by 2 is zero
    IF (Number MODULO 2) IS EQUAL TO 0 THEN
        // 4. If zero, it is even
        PRINT Number + " is Even"
    ELSE
        // 5. Otherwise, it is odd
        PRINT Number + " is Odd"
    END IF
    
    END
```

```python
# Define the number to check
number = 7

# Check if the number is even or odd
if number % 2 == 0:
    print(f"{number} is Even")
else:
    print(f"{number} is Odd")
# Output: 7 is Odd
```

### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Add your code solution here
ALGORITHM SumN_NaturalNumbers
    // 1. Start the process
    START
    
    // 2. Define the limit N
    READ N = 10 
    
    // 3. Initialize the sum variable
    Sum = 0
    
    // 4. Loop from 1 up to N
    FOR i FROM 1 TO N DO
        // 5. Add the current number to the sum
        Sum = Sum + i
    END FOR
    
    // 6. Display the result
    PRINT "The sum of the first " + N + " numbers is: " + Sum
    
    END
```

```python
# Define the limit N
N = 10
total_sum = 0

# Loop from 1 up to N (inclusive)
for i in range(1, N + 1):
    total_sum += i

# Print the result
print(f"The sum of the first {N} numbers is: {total_sum}")
# Output: The sum of the first 10 numbers is: 55
```

### Bonus Exercise 7: Print Multiplication Table of a Number

Write pseudocode to print the multiplication table of a given number up to 10.

<!-- Add your pseudocode here -->

```python
# Add your code solution here
ALGORITHM MultiplicationTable
    // 1. Start the process
    START
    
    // 2. Define the number for the table
    READ Number = 5
    
    // 3. Loop from 1 up to 10
    FOR i FROM 1 TO 10 DO
        // 4. Calculate the product
        Product = Number * i
        
        // 5. Print the multiplication line
        PRINT Number + " x " + i + " = " + Product
    END FOR
    
    END
```

```python
# Define the number for the multiplication table
number = 5

# Loop from 1 to 10
for i in range(1, 11):
    product = number * i
    print(f"{number} x {i} = {product}")
# Output: 5 x 1 = 5, ..., 5 x 10 = 50
```

### Bonus Exercise 9: Find Factorial of a Number

Write pseudocode that calculates the factorial of a given positive integer.

<!-- Add your pseudocode here -->

```python
# Add your code solution here
ALGORITHM Factorial
    // 1. Start the process
    START
    
    // 2. Define the input number
    READ N = 5 
    
    // 3. Initialize the result variable
    FactorialResult = 1
    
    // 4. Handle non-negative constraints
    IF N < 0 THEN
        PRINT "Factorial is undefined for negative numbers"
    ELSE IF N = 0 THEN
        PRINT "The factorial of 0 is 1"
    ELSE
        // 5. Iterate from 1 up to N
        FOR i FROM 1 TO N DO
            // 6. Multiply the result by the current number
            FactorialResult = FactorialResult * i
        END FOR
        // 7. Print the result
        PRINT "The factorial of " + N + " is: " + FactorialResult
    END IF
    
    END
```

```python
# Define the input number
N = 5
factorial_result = 1

if N < 0:
    print("Factorial is undefined for negative numbers")
elif N == 0:
    print("The factorial of 0 is 1")
else:
    # Calculate factorial using a loop
    for i in range(1, N + 1):
        factorial_result *= i
    
    # Print the result
    print(f"The factorial of {N} is: {factorial_result}")
# Output: The factorial of 5 is: 120
```

### Bonus Exercise 10: FizzBuzz Problem

Write pseudocode for the FizzBuzz problem where you print numbers from 1 to N but replace multiples of three with "Fizz" and multiples of five with "Buzz".

<!-- Add your pseudocode here -->

```python
# Add your code solution here
ALGORITHM FizzBuzz
    // 1. Start the process
    START
    
    // 2. Define the limit N
    READ N = 15 
    
    // 3. Iterate from 1 up to N
    FOR i FROM 1 TO N DO
        // 4. Initialize an empty string for the output
        Output = ""
        
        // 5. Check for multiple of 3
        IF (i MODULO 3) IS EQUAL TO 0 THEN
            Output = Output + "Fizz"
        END IF
        
        // 6. Check for multiple of 5
        IF (i MODULO 5) IS EQUAL TO 0 THEN
            Output = Output + "Buzz"
        END IF
        
        // 7. Print the number if output is empty, otherwise print Output
        IF Length(Output) IS EQUAL TO 0 THEN
            PRINT i 
        ELSE
            PRINT Output
        END IF
    END FOR
    
    END
```

```python
# Define the limit N
N = 15

# Iterate from 1 up to N
for i in range(1, N + 1):
    output = ""
    
    if i % 3 == 0:
        output += "Fizz"
        
    if i % 5 == 0:
        output += "Buzz"
        
    # Print the number if output is empty, otherwise print the output string
    if not output:
        print(i)
    else:
        print(output)
# Output: 1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, 14, FizzBuzz
```
